```python
import pandas as pd
import scrapy
from scrapy.crawler import CrawlerProcess
import requests
import json
```

```python
class LeyChile_Spider(scrapy.Spider):
    name = 'leychile_spider'
    start_urls = ['https://www.bcn.cl/leychile/consulta/listaresultadosavanzada?stringBusqueda=-1%23normal%23on%7C%7C2%23normal%23XX1%7C%7C117%23normal%23on%7C%7C48%23normal%23on&tipoNormaBA=&npagina=1&itemsporpagina=10&orden=2&tipoviene=4&totalitems=&seleccionado=0&taxonomia=&valor_taxonomia=&o=experta']
    headers = {
        'accept': 'application/json, text/plain, */*',
        'accept-encoding': 'gzip, deflate, br',
        'accept-language': 'es-419,es;q=0.9,en;q=0.8',
        'cache-control': 'no-cache',
        'origin': 'https://www.bcn.cl',
        'pragma': 'no-cache',
        'referer': 'https://www.bcn.cl/',
        'sec-fetch-dest': 'empty',
        'sec-fetch-mode': 'cors',
        'sec-fetch-site': 'cross-site',
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36'
    }
    
    def parse(self, response):
        url = 'https://nuevo.leychile.cl/servicios/Consulta/getCodigos'
        yield scrapy.Request(url, callback = self.parse_cod, headers=self.headers)
        
    def parse_cod(self, response):
        base_url = 'https://nuevo.leychile.cl/servicios/Navegar/get_norma_json?idNorma='
        raw_data = response.body
        data = json.loads(raw_data)
        for codigo in data:
            codigo_id = codigo['idNorma']
            codigo_url = base_url + codigo_id
            request = scrapy.Request(codigo_url, callback = self.parse_ley, headers = self.headers)
            yield request
            
    def parse_ley(self, response):
        raw_data = response.body
        data = json.loads(raw_data)
        yield {
            'Titulo': data['html']['html']['0']['t'],
            'Articulo_1': data['html']['html']['1']['t']
        }
```

```python
process = CrawlerProcess()
process.crawl(LeyChile_Spider)
process.start()
```
